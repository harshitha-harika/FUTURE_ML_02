```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

# Dataset
data = {
    "Ticket": [
        "System crash issue",
        "Payment failed",
        "Password reset needed",
        "Website loading slowly",
        "Account hacked"
    ],
    
    "Category": [
        "Technical",
        "Billing",
        "Support",
        "Technical",
        "Security"
    ]
}

# Create dataframe
df = pd.DataFrame(data)

# Convert text into numbers
vectorizer = CountVectorizer()

X = vectorizer.fit_transform(df["Ticket"])

y = df["Category"]

# Train model
model = MultinomialNB()

model.fit(X, y)

# Test tickets
new_tickets = [
    "Unable to login account",
    "Refund not received",
    "Website crashed"
]

# Transform new data
new_X = vectorizer.transform(new_tickets)

# Predict
predictions = model.predict(new_X)

# Output
for ticket, category in zip(new_tickets, predictions):
    print("Ticket:", ticket)
    print("Predicted Category:", category)
    print()
```
